```python
import pandas as pd
import numpy as np
```

```python
# Read the training and test data and merge them
df_train = pd.read_csv("./fraudTrain.csv")
df_test = pd.read_csv("./fraudTest.csv")
```

```python
df_train.columns
```

```python
user_columns = ['first', 'last', 'gender', 'cc_num', 'job', 'dob']
df_users = pd.concat([df_train[user_columns], df_test[user_columns]])
```
